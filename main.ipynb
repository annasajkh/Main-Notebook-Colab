```
%cd drive/MyDrive
!mkdir nikos

import twint
import re
import nest_asyncio
import urllib.request
import imagecleaner

nest_asyncio.apply()

urls = []
count = 0;

c = twint.Config()
c.Search = "niko #oneshotgame"
c.Min_likes = 10
c.Images = True
c.Hide_output = True
c.Store_object = True
twint.run.Search(c)

tweets = twint.output.tweets_list

for tweet in tweets:
  for url in tweet.photos:
    urls.append(url)



for url in urls:
  urllib.request.urlretrieve(url, f"nikos/img{count}.png")
  count += 1

!image-cleaner nikos

print(f"downloaded {count + 1} images")


```
